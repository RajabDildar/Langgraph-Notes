This is a simple seqential workflow in langgraph.

```python
from langgraph.graph import  StateGraph, START,END
from typing import TypedDict
```

```python
# Define State
class BMIstate(TypedDict):
    weight_in_kg: float
    height_in_m: float
    bmi:float
    category:str
```

```python
def calculate_bmi(state:BMIstate) -> BMIstate:
    weight = state['weight_in_kg']
    height = state['height_in_m']
    bmi = weight/(height**2)
    state['bmi'] = round(bmi,2)
    return state
```

```python
def label_bmi(state:BMIstate) -> BMIstate:
    bmi = state['bmi']

    if bmi < 18.5:
        state['category'] = "Underweight"
    elif 18.5 <= bmi < 25:
        state['category'] = "Normal"
    elif 25 <= bmi < 30:
        state['category'] = "Overweight"
    else:
        state['category'] = "Obese"
    
    return state
```

```python
# Define graph
graph = StateGraph(BMIstate)

# add nodes to the graph
graph.add_node("calculate_bmi",calculate_bmi)
graph.add_node("label_bmi",label_bmi)

# add edges to the graph
graph.add_edge(START,"calculate_bmi")
graph.add_edge("calculate_bmi","label_bmi")
graph.add_edge("label_bmi",END)

# compile the graph
workflow = graph.compile() # this workflow becomes a runnable behind the scenes. so we can invoke it in execution step.

# execute the graph
```

```python
# execute the graph
initial_state = {"weight_in_kg":100,"height_in_m":1.73}

final_state = workflow.invoke(initial_state)
print(final_state)
```
