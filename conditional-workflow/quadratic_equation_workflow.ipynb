```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict,Literal
```

```python
class QuadState(TypedDict):
    a: int
    b: int
    c: int

    equation: str
    discriminant: float
    result: str
```

```python
def show_equation(state:QuadState):
    eq = f"{state['a']}x2 {state['b']}x {state['c']}"
    return {"equation":eq}

def calc_discriminant(state:QuadState):
    disc = state['b']**2 - (4*state['a']*state['c'])
    return {"discriminant":disc}

def real_roots(state:QuadState):
    root1 = (-state['b'] + state['discriminant'] **0.5) / (2*state['a'])
    root2 = (-state['b'] - state['discriminant'] **0.5) / (2*state['a'])

    result = f"The roots are {root1} and {root2}"
    return {"result":result}

def repeated_roots(state:QuadState):
    root = (-state['b']) / (2*state['a'])
    
    result = f"Only repeating root is: {root}"
    return {"result":result}
    

def no_real_roots(state:QuadState):
    result = "No real roots"
    return {"result":result}

# this function decides which function should be called next.
def check_condition(state:QuadState) -> Literal["real_roots", "repeated_roots","no_real_roots"]:
    if state['discriminant'] > 0:
        return "real_roots"
    elif state["discriminant"] == 0:
        return "repeated_roots"
    else:
        return "no_real_roots"
```

```python
graph = StateGraph(QuadState)

graph.add_node("show_equation",show_equation)
graph.add_node("calc_discriminant",calc_discriminant)
graph.add_node("real_roots",real_roots)
graph.add_node("repeated_roots",repeated_roots)
graph.add_node("no_real_roots",no_real_roots)

graph.add_edge(START,"show_equation")
graph.add_edge("show_equation","calc_discriminant")
graph.add_conditional_edges("calc_discriminant",check_condition) # tell where to start, and the condition checking function
graph.add_edge("real_roots",END)
graph.add_edge("repeated_roots",END)
graph.add_edge("no_real_roots",END)


workflow = graph.compile()
```

```python
initial_state = {
    'a':4,
    'b':2,
    'c':4
}

workflow.invoke(initial_state)
```
